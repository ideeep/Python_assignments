## Question 1
**Why are functions advantageous to have in your programs?**

Functions offer reusability, modularity, abstraction, and testing benefits in programs.

## Question 2
**When does the code in a function run: when it's specified or when it's called?**

The code in a function runs when it's called.

## Question 3
**What statement creates a function?**

The `def` statement creates a function.

## Question 4
**What is the difference between a function and a function call?**

A function is a block of code, while a function call is the execution of that code.

## Question 5
**How many global scopes are there in a Python program? How many local scopes?**

One global scope and multiple local scopes (one for each function call).

## Question 6
**What happens to variables in a local scope when the function call returns?**

Local scope variables are destroyed after the function call returns.

## Question 7
**What is the concept of a return value? Is it possible to have a return value in an expression?**

A return value is what a function sends back to the caller. Yes, return values can be used in expressions.

## Question 8
**If a function does not have a return statement, what is the return value of a call to that function?**

The return value is `None`.

## Question 9
**How do you make a function variable refer to the global variable?**

Use the `global` keyword before the variable name inside the function.

## Question 10
**What is the data type of None?**

The data type of `None` is `NoneType`.

## Question 11
**What does the sentence `import areallyourpetsnamederic` do?**

It raises an `ImportError` as it's not a valid module or package.

## Question 12
**If you had a `bacon()` function in a `spam` module, what would you call it after importing `spam`?**

You would call it using `spam.bacon()`.

## Question 13
**What can you do to save a program from crashing if it encounters an error?**

You can use error handling techniques like `try` and `except`.

## Question 14
**What is the purpose of the `try` clause? What is the purpose of the `except` clause?**

The `try` clause encloses code that might raise exceptions, and the `except` clause specifies how to handle those exceptions.
